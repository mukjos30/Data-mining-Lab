### (f) Create a new qualitative variable, called Elite, by binning the Top10perc variable into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%. Use the value_counts() method of college['Elite'] to see how many elite universities there are. Finally, use the boxplot() method again to produce side-by-side boxplots of Outstate versus Elite.

```python
# Step 1: Create Elite variable
college['Elite'] = pd.cut(
    college2['Top10perc'], 
    bins=[0, 50, 100],      # 0–50 -> 'No', 51–100 -> 'Yes'
    labels=['No', 'Yes']
)

# Step 2: Count how many elite universities there are
college['Elite'].value_counts()
```

```python
# Step 3: Side-by-side boxplots of Outstate versus Elite
college2.boxplot(column='Outstate', by='Elite', figsize=(8,6))
```

###  (g) plot.hist() method of college to produce some histograms with differing numbers of bins for a few of the quantitative variables. The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously. By changing the arguments you can divide the screen up in other combinations.

```python
import matplotlib.pyplot as plt

# Step 1: Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Step 2: Histogram for 'Apps' with 20 bins
college['Apps'].plot.hist(bins=20, ax=axes[0,0], color='skyblue', edgecolor='black')
axes[0,0].set_title('Histogram of Applications')

# Step 3: Histogram for 'Accept' with 15 bins
college['Accept'].plot.hist(bins=15, ax=axes[0,1], color='salmon', edgecolor='black')
axes[0,1].set_title('Histogram of Accepted Students')

# Step 4: Histogram for 'Outstate' with 25 bins
college['Outstate'].plot.hist(bins=25, ax=axes[1,0], color='lightgreen', edgecolor='black')
axes[1,0].set_title('Histogram of Out-of-State Tuition')

# Step 5: Histogram for 'Enroll' with 20 bins
college['Enroll'].plot.hist(bins=20, ax=axes[1,1], color='violet', edgecolor='black')
axes[1,1].set_title('Histogram of Enrollment')

# Step 6: Adjust layout
plt.tight_layout()
plt.show()
```
